```python
def lower_dict(d):
    data_SJR_dict = d.set_index('Title')['SJR'].to_dict()
    new_dict = dict((k.lower(), v) for k, v in data_SJR_dict.items())
    return new_dict
```

```python
data['SJR_2017']= data['SO'].str.lower().map(lower_dict(data2017))
data['SJR_2016']= data['SO'].str.lower().map(lower_dict(data2016))
data['SJR_2015']= data['SO'].str.lower().map(lower_dict(data2015))
data['SJR_2014']= data['SO'].str.lower().map(lower_dict(data2014))
data['SJR_2013']= data['SO'].str.lower().map(lower_dict(data2013))
data['SJR_2012']= data['SO'].str.lower().map(lower_dict(data2012))
data['SJR_2011']= data['SO'].str.lower().map(lower_dict(data2011))
data['SJR_2010']= data['SO'].str.lower().map(lower_dict(data2010))
data['SJR_2009']= data['SO'].str.lower().map(lower_dict(data2009))
data['SJR_2008']= data['SO'].str.lower().map(lower_dict(data2008))
data['SJR_2007']= data['SO'].str.lower().map(lower_dict(data2007))
data['SJR_2006']= data['SO'].str.lower().map(lower_dict(data2006))
data['SJR_2005']= data['SO'].str.lower().map(lower_dict(data2005))
data['SJR_2004']= data['SO'].str.lower().map(lower_dict(data2004))
data['SJR_2003']= data['SO'].str.lower().map(lower_dict(data2003))
data['SJR_2002']= data['SO'].str.lower().map(lower_dict(data2002))
data['SJR_2001']= data['SO'].str.lower().map(lower_dict(data2001))
data['SJR_2000']= data['SO'].str.lower().map(lower_dict(data2000))
data['SJR_1999']= data['SO'].str.lower().map(lower_dict(data1999))
```
