```python
import os
import pandas as pd
import csv
import csv
```

```python
text_dir = 'data/raw_data/2019'
```

```python
files = os.listdir(text_dir) # list all the files 
```

```python
lst = []
for file in files:
    print(file)
    df = pd.read_csv(os.path.join(text_dir,file),sep='\\t',encoding='utf-16')
    lst.append(df)
final_df = pd.concat(lst)
```

```python
final_df.to_csv('2019.csv')
```
