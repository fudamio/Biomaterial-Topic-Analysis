```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw_data/all.csv')
```

```python
df['PY'].value_counts()
```

```python
start = 1973
```

```python
df[df.PY.isin([i for i in range(start,start+4)])].PY.value_counts()
```

```python
for year in range(1973,2017):
    tmp = df[df.PY.isin([i for i in range(year,year+4)])]
    print('Processing {} to {}'.format(year,year+3))
    print(tmp.PY.value_counts())
    tmp.reset_index(drop=True,inplace=True)
    tmp.to_csv('../data/year/{}_to_{}.csv'.format(year,year+3),index=False)
```
