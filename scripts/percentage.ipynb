```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../output/result.csv')
df['Year'] = df['Topic_Year'].apply(lambda x:x.split('_')[1])
df['Topic'] = df['Topic_Year'].apply(lambda x:x.split('_')[0])
```

```python
df.head()
```

```python
df['Growth_Rate'] = df['Topic_num'].pct_change()

df['Citation_Growth_Rate'] = df['Sum_Citation'].pct_change()

df.loc[df['Year'] == '2002','Growth_Rate'] = np.NaN

df.loc[df['Year'] == '2002','Citation_Growth_Rate'] = np.NaN
```

```python
map_dict = df.groupby(['Year'])['Topic_num'].sum().reset_index().set_index('Year').to_dict()
```

```python
map_dict= map_dict['Topic_num']
df['Year_Sum'] = df['Year'].map(map_dict)
```

```python
df['Topic_Percent_By_Year'] = df['Topic_num'] / df['Year_Sum']
```

```python
df['Percent_By_Year_Growth_Rate'] = df['Topic_Percent_By_Year'].pct_change()
df.loc[df['Year'] == '2002','Percent_By_Year_Growth_Rate'] = np.NaN
```

```python
df.to_csv('add_rate.csv',)
```
